```python
import matplotlib.pyplot as plt
import statistics as stats
import numpy as np
import scipy.stats as sp
from scipy.stats import norm
import pandas as pd

```

```python
df = pd.read_csv(r"C:\Users\alana\OneDrive\Documentos\GitHub\Python_E_Mercado_Financeiro\PETR4SA.csv")
```

```python
print(df)
```

# Cálculo do retorno

```python
retorno = (df['Close'].iloc[1:] - df['Close'].iloc[:-1].values) / df['Close'].iloc[:-1].values
```

```python
retorno.head()
```

# Gráfico PETR4 diário

```python
plt.subplot(221)
plt.plot(df.index, df['Close'],color='black')
plt.xlabel('Dias')
plt.title('PETR4 diário (2018-Fev/2023-Mai)')

```

# Gráfico retorno

```python
df_subset = df.iloc[1:, :]
```

```python
plt.subplot(222)
plt.plot(df_subset.index,retorno,color='black')
plt.xlabel('Dias')
plt.title('PETR4 - retorno diário')
```

# Gráfico histograma e dist.normal

```python
plt.subplot(223)
plt.hist(retorno,bins=20,color='gray',density=True)
plt.xlabel('Classes')
plt.ylabel('Frequência')
plt.title('Histograma dos retornos - PETR4')
xmin, xmax = plt.xlim()
media=stats.mean(retorno)
desvio=stats.pstdev(retorno)
eixo_x = np.linspace(xmin,xmax,100)
eixo_y = norm.pdf(eixo_x,media,desvio)
plt.plot(eixo_x,eixo_y,color='black')
```

# Gráfico QQ-Plot  - (Quantile Quantile Plot)

```python
plt.subplot(224)
sp.probplot(retorno,dist="norm", plot=plt)
plt.title('QQ-Plot PETR4')
```
