```python
import matplotlib.pyplot as plt
import statistics as stats
import numpy as np
import scipy.stats as sp
from scipy.stats import norm
from scipy import stats
import pandas as pd

```

```python
df1 = pd.read_csv(r"C:\Users\alana\OneDrive\Documentos\Python_Mercado_Financeiro\^BVSP.csv")
```

```python
df2 = df = pd.read_csv(r"C:\Users\alana\OneDrive\Documentos\GitHub\Python_E_Mercado_Financeiro\PETR4SA.csv")
```

```python
ibovespa_values = df1['Close'].to_numpy()
petr4_values = df2['Close'].to_numpy()
```

```python
print("Tamanho do array ibovespa_values:", len(ibovespa_values))
print("Tamanho do array petr4_values:", len(petr4_values))
```

```python
# Verificar as datas em comum entre os dois conjuntos de dados
common_dates = df1.index.intersection(df2.index)

# Filtrar os arrays apenas para as datas em comum
ibovespa_values = df1.loc[common_dates, 'Close'].to_numpy()
petr4_values = df2.loc[common_dates, 'Close'].to_numpy()

# Verificar os tamanhos dos arrays novamente
print("Tamanho do array ibovespa_values:", len(ibovespa_values))
print("Tamanho do array petr4_values:", len(petr4_values))
```

```python
beta, beta0, r_value, p_value, std_err = stats.linregress(ibovespa_values, petr4_values)
```

```python
print('++++++++Regressão Linear+++++++')
print(' beta        beta0       r_value     p_value     std_err')
yLin = beta*ibovespa_values + beta0
print('%10.6f  %10.6f    %5.2f    %10.6f      %10.8f' %(beta,beta0,r_value,p_value,std_err))
```

```python
plt.plot(ibovespa_values,yLin,'-k',ibovespa_values,petr4_values,'ok')
plt.xlabel('IBOVESPA')
plt.ylabel('PETR4')
```
