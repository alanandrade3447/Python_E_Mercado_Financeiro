```python
import matplotlib.pyplot as plt
import statistics as stats
import numpy as np
import scipy.stats as sp
from scipy.stats import norm
import pandas as pd

```

```python
df = pd.read_csv(r"C:\Users\alana\OneDrive\Documentos\Python_Mercado_Financeiro\^BVSP.csv")
```

```python
print(df)
```

```python
# C치lculo do retorno
```

```python
retorno = (df['Close'].iloc[1:] - df['Close'].iloc[:-1].values) / df['Close'].iloc[:-1].values
```

```python
# Gr치fico histograma e dist.normal
```

```python
plt.subplot(223)
plt.hist(retorno,bins=20,color='gray',density=True)
plt.xlabel('Classes')
plt.ylabel('Frequ칡ncia')
plt.title('Histograma dos retornos - Ibovespa')
xmin, xmax = plt.xlim()
media=stats.mean(retorno)
desvio=stats.pstdev(retorno)
eixo_x = np.linspace(xmin,xmax,100)
eixo_y = norm.pdf(eixo_x,media,desvio)
plt.plot(eixo_x,eixo_y,color='black')
```

```python
# Gr치fico QQ-Plot  - (Quantile Quantile Plot)
```

```python
plt.subplot(224)
sp.probplot(retorno,dist="norm", plot=plt)
plt.title('QQ-Plot Ibovespa')
```
