```python
import matplotlib.pyplot as plt
import statistics as stats
import numpy as np
import scipy.stats as sp
from scipy.stats import norm
from scipy import stats
import pandas as pd

```

```python
df1 = pd.read_csv(r"C:\Users\alana\OneDrive\Documentos\GitHub\Python_E_Mercado_Financeiro\PETR4SA.csv")
```

```python
df2 = pd.read_csv(r"C:\Users\alana\OneDrive\Documentos\GitHub\Python_E_Mercado_Financeiro\VALE3SA.csv")
```

```python
petr4_values = df1['Close'].to_numpy()
vale3_values = df2['Close'].to_numpy()
```

```python
print("Tamanho do array petr4_values:", len(petr4_values))
print("Tamanho do array ibovespa_values:", len(vale3_values))
```

```python
common_dates = df1.index.intersection(df2.index)
petr4_values = df1.loc[common_dates, 'Close'].to_numpy()
vale3_values = df2.loc[common_dates, 'Close'].to_numpy()


print("Tamanho do array petr4_values:", len(petr4_values))
print("Tamanho do array petr4_values:", len(vale3_values))
```

```python
beta, beta0, r_value, p_value, std_err = stats.linregress(petr4_values, vale3_values )
```

```python
print('++++++++Regressão Linear+++++++')
print(' beta        beta0       r_value     p_value     std_err')
yLin = beta*petr4_values + beta0
print('%10.6f  %10.6f    %5.2f    %10.6f      %10.8f' %(beta,beta0,r_value,p_value,std_err))
```

```python
plt.plot(petr4_values,yLin,'-k',petr4_values,vale3_values,'ok')
plt.xlabel('PETR4')
plt.ylabel('VALE3')
```
